```python
fig, ax = plt.subplots()
# count the occurrence of each class 
product_data = data_subset['Product'].value_counts() 
# get x and y data 
points = product_data.index 
frequency = product_data.values 
# create bar chart 
ax.bar(points, frequency)
plt.xticks(rotation=90)
```
